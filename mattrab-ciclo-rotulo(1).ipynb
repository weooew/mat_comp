# Problema de Árvore Geradora Mı́nima com Restrição de Grau Mı́nimo e Centrais Fixos (mdf-MST)
Seja G um grafo conexo, ponderado e ñ direcionado 
G = (V, E) -> conj vértices V, |V| = n, conj arrrestas E, |E| = m
* cada aresta (u,v) tem um peso
* Representação de aresta conectando vértices i e j -> {i,j}
* vértices centrais fixos é um subconj de V -> esses vétices possuem tb grau (= n° de arestas conectadas a esse vértice) mínimo, logo, q° de arrestas conectadas >= dmin do vétice. 
* obj encontrar subárvore de G (= caminho ñ ciclico) que minializa O CUSTO TOTAL e atende às restrições anteriores.    
* não é possível garantir que se encontre uma solução ótima para todas instâncias deste problema em tempo polinomial (tb tem várias restrições de parâmetros de entrada como grafo e grau mínimo dos vértices)

### Árvore Geradora Mínima (MST) 
Encontrar em um grafo ponderado em arestas, uma árvore geradora de custo mínimo.
    * Restrições sobre os graus dos vértices na árvore geradora
Seja G = (V,E) um grafo conexo e não direcionado com |V | = n e |E| = m.
Associe... a cada aresta(e) um custo (ce ≥ 0, custo de arresta) 
e a cada vértice (v) um inteiro positivo d(v), relacionado ao *grau* de v na árvore.


Árvore Geradora Mı́nima com Restrição de Grau Mı́nimo: encontrar uma árvore geradora mı́nima de G, com que cada vértice v tem grau pelo menos d(v) ou é uma folha (grau 1) na árvore.(algoritmo branch-and-bound?)
Outras formulações e resultados computacionais para
md-MST podem ser vistos em (Martinez; Cunha, 2012; Martinez; Cunha, 2013).

Na literatura, normalmente considera-se d(v) = d, ∀v ∈ V , e chama-se os vértices com
grau mı́nimo d de centrais e as folhas, de terminais. Observe que no md-MST os vértices
centrais e terminais, ou mesmo suas quantidades, são arbitrários. Neste trabalho, des-
crevemos uma variação do problema, onde os vértices centrais e terminais são fixados a
priori. Vamos chamá-lo de Problema de Árvore Geradora Mı́nima com Restrição de Grau
Mı́nimo e Centrais Fixos (mdf-MST). A Figura 1 ilustra uma instância do problema para

d = 2
em cada árvore viável(T) o subgrafo de T induzido pelos centrais é uma subárvore. Em uma solução ótima, porém, essa subárvore pode não ser geradora mı́nima para o subgrafo induzido pelos centrais.

### Formulação matemática
Particionamos V = C ∪ T, |C| = c, |T| = t, c + t = n, onde C e T descrevem os conj de centrais e terminais, respectivamente. 
E(C) -> conj de arestas entre os vértices centrais
E(C, T) -> conj de arestas entre centrais e terminais. 

Por simplicidade, consideramos que não há arestas em G ligando dois terminais, já que elas nunca farão parte de uma solução.

O subgrafo induzido pelos vértices centrais será denotado por G(C). 
Consideramos que d(v) = d, ∀v ∈ C, e denotar por dv o grau do vértice v numa solução viável qualquer.
Para que os problemas em estudo possam ser viáveis: supor que o grau, em G, de cada vértice de C é maior ou igual a d e que G(C) é conexo. Também supomos que d ≥ 2, pois o problema se resume ao MST se d = 1.

### Solução por eliminação de ciclos.
#### Imprime uma tabela com resultados dos 3 grafos (com p 0.5, 0.8 e 1). 
#### Assim, comparando a densidade, valor objetivo, tempo em ms e Iterações):

```python
from ortools.linear_solver import pywraplp
import os
import networkx as nx 
import random
import matplotlib.pyplot as plt
from itertools import combinations
import pandas as pd

# Função para ler arquivo
def lerArquivo(arquivo): 
    with open(arquivo, 'r') as file:
        linhas = file.readlines()
    
    # 1ª linha: número de vértices, número de centrais, número de arestas
    nv, nc, na = map(int, linhas[0].split())
    
    centrais = []
    arestas = []
    
    # Linhas seguintes: centrais
    for i in range(1, nc + 1): #4, i = 0
        vertice, grau_minimo = map(int, linhas[i].split())
        centrais.append((vertice, grau_minimo))
    
    # Linhas seguintes: arestas
    for i in range(nc + 1, nc + na + 1):
        u, v, custo = map(int, linhas[i].split())
        arestas.append((u, v, custo))
        
    #criar grafo para esses parametros passados
    G = nx.Graph()  
    G.add_nodes_from(range(1, nv + 1))  # Add vértices
    for u, v, custo in arestas:
        G.add_edge(u, v, weight=custo)  # Add arestas e seus pesos

    #print(f"nv: {nv}")
    #print(f"nc: {nc}")
    #print(f"na: {na}")
    #print(f"centrais: {centrais}c")
    #print(f"arestas: {arestas}a")
        
    return nv, nc, G, centrais, arestas

def gerar_grafo(nv, G, centrais, arestas, p1):
    g = nx.Graph()
    g.add_nodes_from(G.nodes())  # Adicionar os mesmos vértices do grafo original

    while not nx.is_connected(g):
        # Adicionar arestas com probabilidade p1 de existir
        for u, v, custo in arestas:
            if random.random() <= p1:
                g.add_edge(u, v, weight=custo) # Add

    # Visualizar grafo g1
    plt.figure()
    nx.draw(g, with_labels=True)
    plt.title(f"Grafo com p={p1}")
    plt.show()

    return g

# Função para resolver o problema e retirar soluções com ciclos (1ª forma)
# Método Eliminação de ciclos
def eliminarCiclos(nv, nc, G, centrais, arestas):

#    print(f"N°vert: {nv}")
  #  print(f"N°centrais: {nc}")
  #  print(f"Centrais: {centrais}c")
   # print(f"Arestas: {arestas}a")
    
    # Inicializar o solver
    solver = pywraplp.Solver.CreateSolver('SCIP')
    solver.SetTimeLimit(1800 * 1000) #1800 seg

    T = nx.Graph()  # inicializar grafo vazio
    T.add_nodes_from(G.nodes())  # tem mesmas vértices
    if not nx.is_connected(G):
        print("grafo não é conexo")
        return None

    # criar um x binário para cada aresta (u, v)
    x = {}  # dicionario
    for (u, v, custo) in arestas:  # preencher com (u1, v1): x1
        x[(u, v)] = solver.IntVar(0, 1, f'x_{u}_{v}') #.format(u,v)?
        
    arestas_centrais = [(u, v, custo) for (u, v, custo) in arestas if u in [i for i, _ in centrais] and v in [i for i, _ in centrais]]
    
    # Função objetivo: minimizar o custo total das arestas selecionadas
    print("\nDefinindo função obj... ps: 1° forma")
    solver.Minimize(solver.Sum(custo * x[(u, v)] for (u, v, custo) in arestas))

    # Restrição 1: Soma de todas as variáveis x(c,c) = c - 1 (ps: c = número de centrais)
    print("Definindo restrições para centrais...")
    solver.Add(solver.Sum(x[u,v] for (u, v, _) in arestas_centrais) == (nc - 1))
    
    #Debug
    #print(f"Número de arestas possíveis entre centrais: {soma_central}")
    #print(f"Valor esperado para a soma: {nc - 1}")
    #print(f"Total de restrições adicionadas: {solver.NumConstraints()}")
    
    # Restrição 2: evitar ciclo (subtours)
    #criar S, eliminar ciclos -> restrições limita n° de arestas desse S  (n° arestas de S <= |S| - 1)
    #print("Adicionando restrições para evitar ciclos...")
    for size in range(2, nc+1): #size entre 2 e nc - 1 (todos subconj possíveis)
        #print(f"Tamanho: {size}")
        for subconj in combinations(range(1,nc+1), size): #todas combinações com vértices de size 0 a nc
            #(0,1,2,3)
            # i = 0, j assume 1, 2, 3 -> (0,1) (0,2) (0,3)
            # i = 1, j assume 2, 3 -> (1,2) (1,3)
            # i = 2, j assume 3 -> (2,3)    
            #todas combinações possíveis: (0,1) (0,2) (0,3) (1,2) (1,3) (2,3)
            #calcular só as arestas existentes no Grafo original
            #print(f"Subconjunto considerado: {subconj}")
            arestas_subconj = [(u, v) for (u, v, _) in arestas_centrais if u in subconj and v in subconj and u < v] 
            atual = solver.Sum(x[(u, v)] for (u, v) in arestas_subconj if (u, v) in x or (v, u) in x)
            #print(f"Arestas no subconjunto {subconj}: {atual}")
            solver.Add(atual <= size - 1)          
            #print(f" atual <= {size-1}")
            #print(f"Total de restrições adicionadas: {solver.NumConstraints()}")
            
    # Restrição 3: Cada vértice central deve obedecer seu grau mínimo
    print("Definindo restrições de grau mínimo...")
    for i, grau_min in centrais: 
        solver.Add(solver.Sum(x[(u, v)] for (u, v, _) in arestas if u == i or v == i) >= grau_min)
    
    # Restrição 4: todos verts t(ñ centrais) tem exatamente uma aresta entre central e ñ central
    print("Definindo restrições para vértices não centrais...\n")
    for t in range(1, nv+1):
        if t not in [i for i, _ in centrais]:
            solver.Add(solver.Sum(x[(u, v)] for (u, v, custo) in arestas if (u == t and v in [i for i, _ in centrais]) or (v == t and u in [i for i, _ in centrais])) == 1)
    
    # Resolver o modelo
    status = solver.Solve()

    # Preencher o caminho da solução com arestas selecionadas
    if status == pywraplp.Solver.OPTIMAL:
        densidade = nx.density(G)
        caminho_solucao = []
        for (u, v, custo) in arestas:
            if x[(u, v)].solution_value() == 1:
                caminho_solucao.append((u, v, custo))
                T.add_edge(u, v, weight=custo)
    
        valor_objetivo = solver.Objective().Value()
        tempo_resolvido = solver.wall_time()
        iteracoes = solver.iterations()

        return densidade, valor_objetivo, tempo_resolvido, iteracoes

    else:
        print('The problem does not have an optimal solution.')
        return None, None, None, None
        

def tabela(grafos, resultados_eliminacao):
    colunas = ["Grafo", "Densidade", "Valor Objetivo", "Tempo (ms)", "Iterações"]

    # Dados para Eliminação de Ciclos
    data_eliminacao = []
    for i, grafo in enumerate(grafos):
        densidade, valor_objetivo, tempo, iteracoes = resultados_eliminacao[i]
        data_eliminacao.append([f"Grafo g{i+1}", densidade, valor_objetivo, tempo, iteracoes])

    tabela_eliminacao = pd.DataFrame(data_eliminacao, columns=colunas)

    print("\nComparação - Eliminação de Ciclos:")
    print(tabela_eliminacao)


# Ler arquivos e resolver o problema
pasta = 'arqsFabio'
arquivos = [f for f in os.listdir(pasta) if f.endswith('.txt')]

# Exibir opções de arquivos para o usuário
print("Opções:")
for i, x in enumerate(arquivos):
    print(f"{i + 1}: {x}")

# Escolher um arquivo específico
teste = int(input("Escolha o número do arquivo que deseja usar: ")) - 1
arquivo_selecionado = os.path.join(pasta, arquivos[teste])

# Nome do arquivo escolhido
print(f"\nArquivo escolhido: {arquivo_selecionado}\n")
nv, nc, G, centrais, arestas = lerArquivo(arquivo_selecionado)

print("Grafo de g1 com 0.5:")
g1 = gerar_grafo(nv, G, centrais, arestas, 0.5)
print("Grafo de g2 com 0.8:")
g2 = gerar_grafo(nv, G, centrais, arestas, 0.8)
print("Grafo de g2 com 1:")
g3 = gerar_grafo(nv, G, centrais, arestas, 1)

grafos = [g1, g2, g3]

# Resolver para o método de Eliminação de Ciclos
resultados_eliminacao = [
    eliminarCiclos(nv, nc, g, centrais, arestas) for g in grafos
]

# Comparar os resultados
tabela(grafos, resultados_eliminacao)
```

```python
from ortools.linear_solver import pywraplp
import os
import networkx as nx 
import random
import matplotlib.pyplot as plt
from itertools import combinations
import pandas as pd

# Função para ler arquivo
def lerArquivo(arquivo): 
    with open(arquivo, 'r') as file:
        linhas = file.readlines()
    
    # 1ª linha: número de vértices, número de centrais, número de arestas
    nv, nc, na = map(int, linhas[0].split())
    
    centrais = []
    arestas = []
    
    # Linhas seguintes: centrais
    for i in range(1, nc + 1): #4, i = 0
        vertice, grau_minimo = map(int, linhas[i].split())
        centrais.append((vertice, grau_minimo))
    
    # Linhas seguintes: arestas
    for i in range(nc + 1, nc + na + 1):
        u, v, custo = map(int, linhas[i].split())
        arestas.append((u, v, custo))
        
    #criar grafo para esses parametros passados
    G = nx.Graph()  
    G.add_nodes_from(range(1, nv + 1))  # Add vértices
    for u, v, custo in arestas:
        G.add_edge(u, v, weight=custo)  # Add arestas e seus pesos

    #print(f"nv: {nv}")
    #print(f"nc: {nc}")
    #print(f"na: {na}")
    #print(f"centrais: {centrais}c")
    #print(f"arestas: {arestas}a")
        
    return nv, nc, G, centrais, arestas

def gerar_grafo(nv, G, centrais, arestas, p1):
    g = nx.Graph()
    g.add_nodes_from(G.nodes())  # Adicionar os mesmos vértices do grafo original

    while not nx.is_connected(g):
        # Adicionar arestas com probabilidade p1 de existir
        for u, v, custo in arestas:
            if random.random() <= p1:
                g.add_edge(u, v, weight=custo) # Add

    # Visualizar grafo g1
    plt.figure()
    nx.draw(g, with_labels=True)
    plt.title(f"Grafo com p={p1}")
    plt.show()

    return g

# Função para resolver o problema e retirar soluções com ciclos (1ª forma)
# Método Eliminação de ciclos
def eliminarCiclos(nv, nc, G, centrais, arestas):

#    print(f"N°vert: {nv}")
  #  print(f"N°centrais: {nc}")
  #  print(f"Centrais: {centrais}c")
   # print(f"Arestas: {arestas}a")
    
    # Inicializar o solver
    solver = pywraplp.Solver.CreateSolver('SCIP')
    solver.SetTimeLimit(1800 * 1000) #1800 seg

    T = nx.Graph()  # inicializar grafo vazio
    T.add_nodes_from(G.nodes())  # tem mesmas vértices
    if not nx.is_connected(G):
        print("grafo não é conexo")
        return None

    # criar um x binário para cada aresta (u, v)
    x = {}  # dicionario
    for (u, v, custo) in arestas:  # preencher com (u1, v1): x1
        x[(u, v)] = solver.IntVar(0, 1, f'x_{u}_{v}') #.format(u,v)?
        
    arestas_centrais = [(u, v, custo) for (u, v, custo) in arestas if u in [i for i, _ in centrais] and v in [i for i, _ in centrais]]
    
    # Função objetivo: minimizar o custo total das arestas selecionadas
    print("\nDefinindo função obj... ps: 1° forma")
    solver.Minimize(solver.Sum(custo * x[(u, v)] for (u, v, custo) in arestas))

    # Restrição 1: Soma de todas as variáveis x(c,c) = c - 1 (ps: c = número de centrais)
    print("Definindo restrições para centrais...")
    solver.Add(solver.Sum(x[u,v] for (u, v, _) in arestas_centrais) == (nc - 1))
    
    #Debug
    #print(f"Número de arestas possíveis entre centrais: {soma_central}")
    #print(f"Valor esperado para a soma: {nc - 1}")
    #print(f"Total de restrições adicionadas: {solver.NumConstraints()}")
    
    # Restrição 2: evitar ciclo (subtours)
    #criar S, eliminar ciclos -> restrições limita n° de arestas desse S  (n° arestas de S <= |S| - 1)
    #print("Adicionando restrições para evitar ciclos...")
    for size in range(2, nc+1): #size entre 2 e nc - 1 (todos subconj possíveis)
        #print(f"Tamanho: {size}")
        for subconj in combinations(range(1,nc+1), size): #todas combinações com vértices de size 0 a nc
            #(0,1,2,3)
            # i = 0, j assume 1, 2, 3 -> (0,1) (0,2) (0,3)
            # i = 1, j assume 2, 3 -> (1,2) (1,3)
            # i = 2, j assume 3 -> (2,3)    
            #todas combinações possíveis: (0,1) (0,2) (0,3) (1,2) (1,3) (2,3)
            #calcular só as arestas existentes no Grafo original
            #print(f"Subconjunto considerado: {subconj}")
            arestas_subconj = [(u, v) for (u, v, _) in arestas_centrais if u in subconj and v in subconj and u < v] 
            atual = solver.Sum(x[(u, v)] for (u, v) in arestas_subconj if (u, v) in x or (v, u) in x)
            #print(f"Arestas no subconjunto {subconj}: {atual}")
            solver.Add(atual <= size - 1)          
            #print(f" atual <= {size-1}")
            #print(f"Total de restrições adicionadas: {solver.NumConstraints()}")
            
    # Restrição 3: Cada vértice central deve obedecer seu grau mínimo
    print("Definindo restrições de grau mínimo...")
    for i, grau_min in centrais: 
        solver.Add(solver.Sum(x[(u, v)] for (u, v, _) in arestas if u == i or v == i) >= grau_min)
    
    # Restrição 4: todos verts t(ñ centrais) tem exatamente uma aresta entre central e ñ central
    print("Definindo restrições para vértices não centrais...\n")
    for t in range(1, nv+1):
        if t not in [i for i, _ in centrais]:
            solver.Add(solver.Sum(x[(u, v)] for (u, v, custo) in arestas if (u == t and v in [i for i, _ in centrais]) or (v == t and u in [i for i, _ in centrais])) == 1)
    
    # Resolver o modelo
    status = solver.Solve()

    # Preencher o caminho da solução com arestas selecionadas
    if status == pywraplp.Solver.OPTIMAL:
        densidade = nx.density(G)
        caminho_solucao = []
        for (u, v, custo) in arestas:
            if x[(u, v)].solution_value() == 1:
                caminho_solucao.append((u, v, custo))
                T.add_edge(u, v, weight=custo)
    
        valor_objetivo = solver.Objective().Value()
        tempo_resolvido = solver.wall_time()
        iteracoes = solver.iterations()

        return densidade, valor_objetivo, tempo_resolvido, iteracoes

    else:
        print('The problem does not have an optimal solution.')
        return None, None, None, None
        
    # Função para resolver o problema com rótulos nos vértices (2ª forma)
def solucao_rotulos_vertices(nv, nc, G, centrais, arestas):    
    # Inicializar o pywraplp solver
    solver = pywraplp.Solver.CreateSolver('SCIP')
    solver.SetTimeLimit(1800 * 1000) #1800 seg

    # Inicializar grafo auxiliar com os mesmos vértices de G
    H = nx.Graph()
    H.add_nodes_from(G.nodes())
    if not nx.is_connected(G):
        print("Graph given is not connected.")
        return None
    
    # Escolher aleatoriamente o vértice raíz "r" entre os centrais
    ind_r = random.randrange(nc)
    r, r_grau_min = centrais[ind_r]
    centrais_sem_raiz = centrais[:ind_r] + centrais[ind_r+1:]
    
    # Conjuntos de vértices V\{r}, C e C\{r}
    vertices_sem_raiz = list(range(1, r)) + list(range(r+1, nv+1))
    vertices_centrais = [i for i, _ in centrais]
    vertices_centrais_sem_raiz = [i for i, _ in centrais_sem_raiz]
    
    arcos = [] # arcos (i, j), excluindo arcos partindo dos vértices de C para r e partindo dos vértices de T
    y = {}  # y binário para cada um desses arcos
    delta_saindo = {i : [] for i in vertices_centrais} # Conjunto de todos os arcos que saem do vértice i
    delta_chegando = {i : [] for i in vertices_sem_raiz} # Conjunto de todos os arcos que chegam ao vértice i
    for i, j, custo in arestas:
        if i in vertices_centrais and j in vertices_sem_raiz:
            arcos.append((i, j, custo))
            y[(i, j)] = solver.IntVar(0, 1, f'y_{i}_{j}')
            delta_saindo[i].append(j)
            delta_chegando[j].append(i)
        if j in vertices_centrais and i in vertices_sem_raiz:
            arcos.append((j, i, custo))
            y[(j, i)] = solver.IntVar(0, 1, f'y_{j}_{i}')
            delta_saindo[j].append(i)
            delta_chegando[i].append(j)
    
    # Função objetivo: minimizar o custo total dos arcos selecionados
    print("\nDefinindo função obj... ps: 2° forma")
    solver.Minimize(solver.Sum(custo * y[(u, v)] for (u, v, custo) in arcos))
    
    # Restrição 1: Soma dos arcos centrais "chegando" em j = 1, para todo vértice j diferente da raíz 
    print("Definindo restrições aos vértices diferentes da raíz...")
    for j in vertices_sem_raiz:
        solver.Add(solver.Sum(y[(i, j)] for i in delta_chegando[j]) == 1)

    # Restrições 2 & 3: Cada vértice central deve obedecer seu grau mínimo
    print("Definindo restrições de grau mínimo aos vértices centrais...")
    for i, grau_min in centrais_sem_raiz:
        solver.Add(solver.Sum(y[(i, j)] for j in delta_saindo[i]) >= grau_min - 1)
    
    solver.Add(solver.Sum(y[(r, j)] for j in delta_saindo[r]) >= r_grau_min)
    
    # Restrição 4, 5, & 6: Atribuir rótulos aos vértices com condiçõi%ies, de modo a evitar a formação de ciclos entre os arcos escolhidos
    print("Definindo restrições aos rótulos dos vértices...")
    
    # Criar rótulo para cada vértice central
    u = {i : solver.IntVar(0, nc, f'u_{i}') for i in range(1, nc+1)}

    # Restrição de arcos e rótulos onde cada arco na árvore é direcionado de um vértice com menor rótulo para um de maior
    for i, j, _ in arcos:
        if j in vertices_centrais_sem_raiz:
            solver.Add(u[i] - u[j] + nc * y[(i, j)] <= nc - 1)
    
    # Condições de rótulos aos vértices centrais diferentes da raíz
    for i in vertices_centrais_sem_raiz:
        solver.Add(u[i] >= 1)
        solver.Add(u[i] <= nc - 1)
    
    # Rótulo 0 atribuido à raíz
    solver.Add(u[r] == 0)
    
    # Resolver o modelo
    print("Resolvendo modelo...")
    status = solver.Solve()

    # Preencher o caminho da solução com arestas selecionadas
    if status == pywraplp.Solver.OPTIMAL:
        densidade = nx.density(G)
        caminho_solucao = []
        for (i, j, custo) in arcos:
            if y[(i, j)].solution_value() == 1:
                caminho_solucao.append((i, j, custo))
                H.add_edge(i, j, weight=custo)
    
        valor_objetivo = solver.Objective().Value()
        tempo_resolvido = solver.wall_time()
        iteracoes = solver.iterations()

        return densidade, valor_objetivo, tempo_resolvido, iteracoes

    else:
        print('The problem does not have an optimal solution.')
        return None, None, None, None
        
def tabela(grafos, resultados_rotulos):
    colunas = ["Grafo", "Densidade", "Valor Objetivo", "Tempo (ms)", "Iterações"]
    
    # Dados para Rótulos nos Vértices
    data_rotulos = []
    for i, grafo in enumerate(grafos):
        densidade, valor_objetivo, tempo, iteracoes = resultados_rotulos[i]
        data_rotulos.append([f"Grafo g{i+1}", densidade, valor_objetivo, tempo, iteracoes])

    tabela_rotulos = pd.DataFrame(data_rotulos, columns=colunas)

    print("\nComparação - Rótulos nos Vértices:")
    print(tabela_rotulos)

# Ler arquivos e resolver o problema
pasta = 'arqsFabio'
arquivos = [f for f in os.listdir(pasta) if f.endswith('.txt')]

# Exibir opções de arquivos para o usuário
print("Opções:")
for i, x in enumerate(arquivos):
    print(f"{i + 1}: {x}")

# Escolher um arquivo específico
teste = int(input("Escolha o número do arquivo que deseja usar: ")) - 1
arquivo_selecionado = os.path.join(pasta, arquivos[teste])

# Nome do arquivo escolhido
print(f"\nArquivo escolhido: {arquivo_selecionado}\n")
nv, nc, G, centrais, arestas = lerArquivo(arquivo_selecionado)

print("Grafo de g1 com 0.5:")
g1 = gerar_grafo(nv, G, centrais, arestas, 0.5)
print("Grafo de g2 com 0.8:")
g2 = gerar_grafo(nv, G, centrais, arestas, 0.8)
print("Grafo de g2 com 1:")
g3 = gerar_grafo(nv, G, centrais, arestas, 1)

grafos = [g1, g2, g3]

# Resolver para o método de Rótulos nos Vértices
resultados_rotulos = [
    solucao_rotulos_vertices(nv, nc, g, centrais, arestas) for g in grafos
]

# Comparar os resultados
tabela(grafos, resultados_rotulos)
```

```python
from ortools.linear_solver import pywraplp
import os
import networkx as nx 
import random
import matplotlib.pyplot as plt
from itertools import combinations
import pandas as pd

# Função para ler arquivo
def lerArquivo(arquivo): 
    with open(arquivo, 'r') as file:
        linhas = file.readlines()
    
    # 1ª linha: número de vértices, número de centrais, número de arestas
    nv, nc, na = map(int, linhas[0].split())
    
    centrais = []
    arestas = []
    
    # Linhas seguintes: centrais
    for i in range(1, nc + 1): #4, i = 0
        vertice, grau_minimo = map(int, linhas[i].split())
        centrais.append((vertice, grau_minimo))
    
    # Linhas seguintes: arestas
    for i in range(nc + 1, nc + na + 1):
        u, v, custo = map(int, linhas[i].split())
        arestas.append((u, v, custo))
        
    #criar grafo para esses parametros passados
    G = nx.Graph()  
    G.add_nodes_from(range(1, nv + 1))  # Add vértices
    for u, v, custo in arestas:
        G.add_edge(u, v, weight=custo)  # Add arestas e seus pesos

    #print(f"nv: {nv}")
    #print(f"nc: {nc}")
    #print(f"na: {na}")
    #print(f"centrais: {centrais}c")
    #print(f"arestas: {arestas}a")
        
    return nv, nc, G, centrais, arestas

def gerar_grafo(nv, G, centrais, arestas, p1):
    g = nx.Graph()
    g.add_nodes_from(G.nodes())  # Adicionar os mesmos vértices do grafo original

    while not nx.is_connected(g):
        # Adicionar arestas com probabilidade p1 de existir
        for u, v, custo in arestas:
            if random.random() <= p1:
                g.add_edge(u, v, weight=custo) # Add

    # Visualizar grafo g1
    plt.figure()
    nx.draw(g, with_labels=True)
    plt.title(f"Grafo com p={p1}")
    plt.show()

    return g
    
# Função para resolver o problema e retirar soluções com ciclos (1ª forma)
# Método Eliminação de ciclos
def eliminarCiclos(nv, nc, G, centrais, arestas):

#    print(f"N°vert: {nv}")
  #  print(f"N°centrais: {nc}")
  #  print(f"Centrais: {centrais}c")
   # print(f"Arestas: {arestas}a")
    
    # Inicializar o solver
    solver = pywraplp.Solver.CreateSolver('SCIP')
    solver.SetTimeLimit(1800 * 1000) #1800 seg

    T = nx.Graph()  # inicializar grafo vazio
    T.add_nodes_from(G.nodes())  # tem mesmas vértices
    if not nx.is_connected(G):
        print("grafo não é conexo")
        return None

    # criar um x binário para cada aresta (u, v)
    x = {}  # dicionario
    for (u, v, custo) in arestas:  # preencher com (u1, v1): x1
        x[(u, v)] = solver.IntVar(0, 1, f'x_{u}_{v}') #.format(u,v)?
        
    arestas_centrais = [(u, v, custo) for (u, v, custo) in arestas if u in [i for i, _ in centrais] and v in [i for i, _ in centrais]]
    
    # Função objetivo: minimizar o custo total das arestas selecionadas
    print("\nDefinindo função obj... ps: 1° forma")
    solver.Minimize(solver.Sum(custo * x[(u, v)] for (u, v, custo) in arestas))

    # Restrição 1: Soma de todas as variáveis x(c,c) = c - 1 (ps: c = número de centrais)
    print("Definindo restrições para centrais...")
    solver.Add(solver.Sum(x[u,v] for (u, v, _) in arestas_centrais) == (nc - 1))
    
    #Debug
    #print(f"Número de arestas possíveis entre centrais: {soma_central}")
    #print(f"Valor esperado para a soma: {nc - 1}")
    #print(f"Total de restrições adicionadas: {solver.NumConstraints()}")
    
    # Restrição 2: evitar ciclo (subtours)
    #criar S, eliminar ciclos -> restrições limita n° de arestas desse S  (n° arestas de S <= |S| - 1)
    #print("Adicionando restrições para evitar ciclos...")
    for size in range(2, nc+1): #size entre 2 e nc - 1 (todos subconj possíveis)
        #print(f"Tamanho: {size}")
        for subconj in combinations(range(1,nc+1), size): #todas combinações com vértices de size 0 a nc
            #(0,1,2,3)
            # i = 0, j assume 1, 2, 3 -> (0,1) (0,2) (0,3)
            # i = 1, j assume 2, 3 -> (1,2) (1,3)
            # i = 2, j assume 3 -> (2,3)    
            #todas combinações possíveis: (0,1) (0,2) (0,3) (1,2) (1,3) (2,3)
            #calcular só as arestas existentes no Grafo original
            #print(f"Subconjunto considerado: {subconj}")
            arestas_subconj = [(u, v) for (u, v, _) in arestas_centrais if u in subconj and v in subconj and u < v] 
            atual = solver.Sum(x[(u, v)] for (u, v) in arestas_subconj if (u, v) in x or (v, u) in x)
            #print(f"Arestas no subconjunto {subconj}: {atual}")
            solver.Add(atual <= size - 1)          
            #print(f" atual <= {size-1}")
            #print(f"Total de restrições adicionadas: {solver.NumConstraints()}")
            
    # Restrição 3: Cada vértice central deve obedecer seu grau mínimo
    print("Definindo restrições de grau mínimo...")
    for i, grau_min in centrais: 
        solver.Add(solver.Sum(x[(u, v)] for (u, v, _) in arestas if u == i or v == i) >= grau_min)
    
    # Restrição 4: todos verts t(ñ centrais) tem exatamente uma aresta entre central e ñ central
    print("Definindo restrições para vértices não centrais...\n")
    for t in range(1, nv+1):
        if t not in [i for i, _ in centrais]:
            solver.Add(solver.Sum(x[(u, v)] for (u, v, custo) in arestas if (u == t and v in [i for i, _ in centrais]) or (v == t and u in [i for i, _ in centrais])) == 1)
    
    # Resolver o modelo
    status = solver.Solve()

    # Preencher o caminho da solução com arestas selecionadas
    if status == pywraplp.Solver.OPTIMAL:
        densidade = nx.density(G)
        caminho_solucao = []
        for (u, v, custo) in arestas:
            if x[(u, v)].solution_value() == 1:
                caminho_solucao.append((u, v, custo))
                T.add_edge(u, v, weight=custo)
    
        valor_objetivo = solver.Objective().Value()
        tempo_resolvido = solver.wall_time()
        iteracoes = solver.iterations()

        return densidade, valor_objetivo, tempo_resolvido, iteracoes

    else:
        print('The problem does not have an optimal solution.')
        return None, None, None, None
        
    # Função para resolver o problema com rótulos nos vértices (2ª forma)
def solucao_rotulos_vertices(nv, nc, G, centrais, arestas):    
    # Inicializar o pywraplp solver
    solver = pywraplp.Solver.CreateSolver('SCIP')
    solver.SetTimeLimit(1800 * 1000) #1800 seg

    # Inicializar grafo auxiliar com os mesmos vértices de G
    H = nx.Graph()
    H.add_nodes_from(G.nodes())
    if not nx.is_connected(G):
        print("Graph given is not connected.")
        return None
    
    # Escolher aleatoriamente o vértice raíz "r" entre os centrais
    ind_r = random.randrange(nc)
    r, r_grau_min = centrais[ind_r]
    centrais_sem_raiz = centrais[:ind_r] + centrais[ind_r+1:]
    
    # Conjuntos de vértices V\{r}, C e C\{r}
    vertices_sem_raiz = list(range(1, r)) + list(range(r+1, nv+1))
    vertices_centrais = [i for i, _ in centrais]
    vertices_centrais_sem_raiz = [i for i, _ in centrais_sem_raiz]
    
    arcos = [] # arcos (i, j), excluindo arcos partindo dos vértices de C para r e partindo dos vértices de T
    y = {}  # y binário para cada um desses arcos
    delta_saindo = {i : [] for i in vertices_centrais} # Conjunto de todos os arcos que saem do vértice i
    delta_chegando = {i : [] for i in vertices_sem_raiz} # Conjunto de todos os arcos que chegam ao vértice i
    for i, j, custo in arestas:
        if i in vertices_centrais and j in vertices_sem_raiz:
            arcos.append((i, j, custo))
            y[(i, j)] = solver.IntVar(0, 1, f'y_{i}_{j}')
            delta_saindo[i].append(j)
            delta_chegando[j].append(i)
        if j in vertices_centrais and i in vertices_sem_raiz:
            arcos.append((j, i, custo))
            y[(j, i)] = solver.IntVar(0, 1, f'y_{j}_{i}')
            delta_saindo[j].append(i)
            delta_chegando[i].append(j)
    
    # Função objetivo: minimizar o custo total dos arcos selecionados
    print("\nDefinindo função obj... ps: 2° forma")
    solver.Minimize(solver.Sum(custo * y[(u, v)] for (u, v, custo) in arcos))
    
    # Restrição 1: Soma dos arcos centrais "chegando" em j = 1, para todo vértice j diferente da raíz 
    print("Definindo restrições aos vértices diferentes da raíz...")
    for j in vertices_sem_raiz:
        solver.Add(solver.Sum(y[(i, j)] for i in delta_chegando[j]) == 1)

    # Restrições 2 & 3: Cada vértice central deve obedecer seu grau mínimo
    print("Definindo restrições de grau mínimo aos vértices centrais...")
    for i, grau_min in centrais_sem_raiz:
        solver.Add(solver.Sum(y[(i, j)] for j in delta_saindo[i]) >= grau_min - 1)
    
    solver.Add(solver.Sum(y[(r, j)] for j in delta_saindo[r]) >= r_grau_min)
    
    # Restrição 4, 5, & 6: Atribuir rótulos aos vértices com condiçõi%ies, de modo a evitar a formação de ciclos entre os arcos escolhidos
    print("Definindo restrições aos rótulos dos vértices...")
    
    # Criar rótulo para cada vértice central
    u = {i : solver.IntVar(0, nc, f'u_{i}') for i in range(1, nc+1)}

    # Restrição de arcos e rótulos onde cada arco na árvore é direcionado de um vértice com menor rótulo para um de maior
    for i, j, _ in arcos:
        if j in vertices_centrais_sem_raiz:
            solver.Add(u[i] - u[j] + nc * y[(i, j)] <= nc - 1)
    
    # Condições de rótulos aos vértices centrais diferentes da raíz
    for i in vertices_centrais_sem_raiz:
        solver.Add(u[i] >= 1)
        solver.Add(u[i] <= nc - 1)
    
    # Rótulo 0 atribuido à raíz
    solver.Add(u[r] == 0)
    
    # Resolver o modelo
    print("Resolvendo modelo...")
    status = solver.Solve()

    # Preencher o caminho da solução com arestas selecionadas
    if status == pywraplp.Solver.OPTIMAL:
        densidade = nx.density(G)
        caminho_solucao = []
        for (i, j, custo) in arcos:
            if y[(i, j)].solution_value() == 1:
                caminho_solucao.append((i, j, custo))
                H.add_edge(i, j, weight=custo)
    
        valor_objetivo = solver.Objective().Value()
        tempo_resolvido = solver.wall_time()
        iteracoes = solver.iterations()

        return densidade, valor_objetivo, tempo_resolvido, iteracoes

    else:
        print('The problem does not have an optimal solution.')
        return None, None, None, None
        
def comparar_metodos(grafos, resultados_eliminacao, resultados_rotulos):
    colunas = ["Grafo", "Densidade", "Valor Objetivo", "Tempo (ms)", "Iterações"]

    # Dados para Eliminação de Ciclos
    data_eliminacao = []
    for i, grafo in enumerate(grafos):
        densidade, valor_objetivo, tempo, iteracoes = resultados_eliminacao[i]
        data_eliminacao.append([f"Grafo g{i+1}", densidade, valor_objetivo, tempo, iteracoes])

    # Dados para Rótulos nos Vértices
    data_rotulos = []
    for i, grafo in enumerate(grafos):
        densidade, valor_objetivo, tempo, iteracoes = resultados_rotulos[i]
        data_rotulos.append([f"Grafo g{i+1}", densidade, valor_objetivo, tempo, iteracoes])

    # Criar as tabelas
    tabela_eliminacao = pd.DataFrame(data_eliminacao, columns=colunas)
    tabela_rotulos = pd.DataFrame(data_rotulos, columns=colunas)

    print("\nComparação - Eliminação de Ciclos:")
    print(tabela_eliminacao)

    print("\nComparação - Rótulos nos Vértices:")
    print(tabela_rotulos)

# Ler arquivos e resolver o problema
pasta = 'arqsFabio'
arquivos = [f for f in os.listdir(pasta) if f.endswith('.txt')]

# Exibir opções de arquivos para o usuário
print("Opções:")
for i, x in enumerate(arquivos):
    print(f"{i + 1}: {x}")

# Escolher um arquivo específico
teste = int(input("Escolha o número do arquivo que deseja usar: ")) - 1
arquivo_selecionado = os.path.join(pasta, arquivos[teste])

# Nome do arquivo escolhido
print(f"\nArquivo escolhido: {arquivo_selecionado}\n")
nv, nc, G, centrais, arestas = lerArquivo(arquivo_selecionado)

print("Grafo de g1 com 0.5:")
g1 = gerar_grafo(nv, G, centrais, arestas, 0.5)
print("Grafo de g2 com 0.8:")
g2 = gerar_grafo(nv, G, centrais, arestas, 0.8)
print("Grafo de g2 com 1:")
g3 = gerar_grafo(nv, G, centrais, arestas, 1)

grafos = [g1, g2, g3]

# Resolver para o método de Eliminação de Ciclos
resultados_eliminacao = [
    eliminarCiclos(nv, nc, g, centrais, arestas) for g in grafos
]

# Resolver para o método de Rótulos nos Vértices
resultados_rotulos = [
    solucao_rotulos_vertices(nv, nc, g, centrais, arestas) for g in grafos
]

# Comparar os resultados
comparar_metodos(grafos, resultados_eliminacao, resultados_rotulos)
```
